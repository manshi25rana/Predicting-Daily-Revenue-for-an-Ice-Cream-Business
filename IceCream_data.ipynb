##Step_04
###Creating Testing and Training Dataset

```python
y = IceCream['Revenue']
```

```python
X = IceCream[['Temperature']]
```

```python
X
```

```python
from sklearn.model_selection import train_test_split
# Imports the train_test_split function used for splitting data into training and testing sets.
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25)
# Spliting the data into training and testing sets (75% train, 25% test).
```

##Step_05
###Train the Model

```python
X_train.shape # Returns the dimensions of the X_train DataFrame (rows, columns).
```

```python
from sklearn.linear_model import LinearRegression
# Imports the LinearRegression class for creating and training a linear regression model.
```

```python
regressor = LinearRegression(fit_intercept=True)
# Creates a LinearRegression object with fit_intercept set to True
# (allowing the model to calculate the intercept).
```

```python
regressor.fit(X_train, y_train)
# Trains the linear regression model using the training data (X_train, y_train).
```

```python
print('Linear Model Coefficient (m): ', regressor.coef_)
print('Linear Model Coefficient (b): ', regressor.intercept_)
```
